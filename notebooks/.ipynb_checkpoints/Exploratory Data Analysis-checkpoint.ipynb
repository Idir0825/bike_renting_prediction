## Analysis

<b>dteday</b> : Will be dropped since it does not bring any information that we don't already know. 
<br>
<b>casual</b> and <b>registered</b>: must be dropped. Our purpose being the prediction of the number of users these columns represent data leakage.
<br>
<b>weathersit</b>, <b>season</b>: we will replace these columns with one hot encoding to avoid giving too much weight to high values.
<br>
<b>weekday</b>, <b>hr</b>, <b>mnth</b>: we will replace these columns with a cyclic encoding (sin, cos). This preserves the periodic nature of these features (e.g. hour 23 being close to hour 0) while avoiding an artificial ordinal relationship and excessive feature dimensionality introduced by one-hot encoding.

## Data cleaning

```python
daily_y = daily_df["cnt"]
hourly_y = hourly_df["cnt"]

daily_df.drop("cnt", axis=1, inplace=True)
hourly_df.drop("cnt", axis=1, inplace=True)
```

```python
columns_to_drop = ["dteday", "casual", "registered"]

for col_to_drop in columns_to_drop:
    if col_to_drop in daily_df.columns : daily_df.drop(col_to_drop, axis=1, inplace=True)
    if col_to_drop in hourly_df.columns : hourly_df.drop(col_to_drop, axis=1, inplace=True)
```

```python
weather_categories = {
    1: "very_good_weather",
    2: "good_weather",
    3: "bad_weather", 
    4: "very_bad_weather"
}

season_categories = {
    1: "spring",
    2: "summer",
    3: "fall",
    4: "winter"
}
```

```python
from sklearn.preprocessing import OneHotEncoder

def one_hot_encode(df: pd.DataFrame,
                  columns: list[str]) -> tuple[pd.DataFrame, OneHotEncoder]:
    """
    One-hot encode specified categorical columns and return the transformed dataframe and the fitted encoder
    """
    encoder = OneHotEncoder(
        sparse_output=False,
        handle_unknown="ignore"
    )

    encoded_array = encoder.fit_transform(df[columns])

    encoded_df = pd.DataFrame(
        encoded_array,
        columns=encoder.get_feature_names_out(columns),
        index=df.index
    )

    df_out = df.drop(columns=columns).join(encoded_df)

    return df_out, encoder
```

```python
categorical_features = ["weathersit", "season"]

daily_df_cat_encoded, daily_df_cat_encoder = one_hot_encode(daily_df, categorical_features)
hourly_df_cat_encoded, hourly_df_cat_encoder = one_hot_encode(hourly_df, categorical_features)
```

```python
display(daily_df_cat_encoded.head(5))
print()
display(hourly_df_cat_encoded.head(5))
```

```python
column_names_map = {f"weathersit_{k}": v for k, v in weather_categories.items()}
column_names_map.update({f"season_{k}": v for k, v in season_categories.items()})
print(column_names_map)
for df in [daily_df_cat_encoded, hourly_df_cat_encoded]:
    df.rename(columns=column_names_map, inplace=True)
```

```python
display(daily_df_cat_encoded.head(5))
print()
display(hourly_df_cat_encoded.head(5))
```

```python
import numpy as np 
def cyclic_encode(df: pd.DataFrame, periods: dict[str, int], drop: bool = True, offsets: dict[str, int] | None = None) -> pd.DataFrame:
    """
    Add cyclic (sin/cos) encoding for specified columns.

    Args:
        df: Input dataframe.
        periods: Mapping column -> period (e.g., {"hr": 24, "weekday": 7, "mnth": 12}).
        drop: If True, drop original columns after encoding.
        offsets: Optional mapping column -> offset to apply before encoding (e.g., {"mnth": 1} for 1..12).

    Returns:
        Transformed dataframe (copy).
    """
    df_out = df.copy()
    offsets = offsets or {}

    for col, period in periods.items():

        if col not in df.columns:
            raise KeyError(f"Column {col} not in dataframe ...")
            
        offset = offsets.get(col, 0)
        x = (df_out[col].astype(float) - float(offset) % float(period))

        angle = 2.0 * np.pi * x / float(period)
        df_out[f"{col}_sin"] = np.sin(angle)
        df_out[f"{col}_cos"] = np.cos(angle)

    if drop:
        df_out = df_out.drop(columns=list(periods.keys()))

    return df_out
```

```python
daily_df_encoded = cyclic_encode(
    daily_df_cat_encoded,
    periods={"weekday": 7, "mnth": 12},
    offsets={"mnth": 1},  # mnth is 1..12 in this dataset
)

hourly_df_encoded = cyclic_encode(
    hourly_df_cat_encoded,
    periods={"hr": 24, "weekday": 7, "mnth": 12},
    offsets={"mnth": 1},  # mnth is 1..12 in this dataset
)
```
